**Problem 1**: 7 pts- Another very important class of special functions are called the _Legendre Polynomials_, say $L_{n}(x)$, which are defined for non-negative integer $n$ via the series

$$
L_{n}(x) = \sum_{k=0}^{n}\frac{(n+k)!}{(k!)^{2}(n-k)!}\left(\frac{x-1}{2}\right)^{k}, ~ -1\leq x \leq 1.
$$

Using a `for`-loop, vectorized, recursive approach, write a function which computes the Legendre functions for any non-negative integer $n$ (4pts).  Make a plot showing $L_{n}(x)$ for $n=0,1,2,3,4$ simultaneously, using different colors for each different $n$.  The plot should have a legend as well as appropriately labeled axes.  (3pts)    

_Answer to Problem 1_

It is given that 

$$
L_{n}(x) = \sum_{k=0}^{n}\frac{(n+k)!}{(k!)^{2}(n-k)!}\left(\frac{x-1}{2}\right)^{k}, ~ -1\leq x \leq 1.
$$

So 

$$
L_{0}(x) = \sum_{k=0}^{0}\frac{(0+0)!}{(0!)^{2}(0)!}\left(\frac{0-1}{2}\right)^{0} = 1, ~ -1\leq x \leq 1.
$$

Now, 
Lets define the function using python:

```python
def factorial(n): 
    return 1 if (n==1 or n==0) else n * factorial(n - 1)  
def Legendre_poly(n, x):
    tot = np.ones(x.size)
    ak = np.ones(x.size)
    if n>0:
        for k in range(1, int(n)+1):
            top = (factorial(n+k))*(((x-1.)/2.)**k)
            bottom = ((factorial(k))**2)*(factorial(n-k))
            ak = top/bottom
            tot += ak
        return tot
    else:
        return tot
```

```python
xvals = np.linspace(-1, 1, int(1e5))
yvals1 = Legendre_poly(0,xvals)
yvals2 = Legendre_poly(1,xvals)
yvals3 = Legendre_poly(2,xvals)
yvals4 = Legendre_poly(3,xvals)
yvals5 = Legendre_poly(4,xvals)

plt.plot(xvals, yvals1, color = 'b', label = '$L_{0}(x)$')
plt.plot(xvals, yvals2, color = 'g', label = '$L_{1}(x)$')
plt.plot(xvals, yvals3, color = 'r', label = '$L_{2}(x)$')
plt.plot(xvals, yvals4, color = 'c', label = '$L_{3}(x)$')
plt.plot(xvals, yvals5, color = 'm', label = '$L_{4}(x)$')


plt.title("Legendre Polynomial Function")
plt.xlabel("x")
plt.ylabel("$L_{n}(x)$")
plt.legend()
```

We can double check this graphs by plugging in n for the Legendre Polynomial, and we get:
1. $L_0= 1$
2. $L_1 = x$ 
Which is both visible in the graph above. 

**Problem 2**: 5 pts - Using a for loop based approach, write a program which finds the function $f(n)$ where 

$$
f(n) = \frac{1}{n}\sum_{j=1}^{n} \left(\left(\frac{j}{n}\right)^{5} + 6\left(\frac{j}{n}\right)^{3} \right)
$$

for any positive integer n (2pts).  Generate a labeled plot of the results of your function for $1\leq n \leq 1000$, where $n$ will be on the $x$-axis, while the results of your function will appear on the $y$-axis.  Describe any trends you see from your graph. (3pts)      

_Answer to Problem 2_

```python
def function_p2(n):
    tot = 0
    for jj in range(1, int(n)+1):
            ak = ((jj/n)**5.) + 6.*((jj/n)**3.)
            tot += ak
    return tot/n
```

```python
nvals = np.linspace(1., 1000., 1000)
yvals = [function_p2(nval) for nval in nvals]

plt.plot(nvals, yvals)
plt.title("f(n)")
plt.xlabel("n")
plt.ylabel("$f(n)$")
plt.legend()
```

In the graph above it is visible that f(n) is converging to a certain point as n gets bigger. In fact when n=1000, function_p2 = 1.6701685833332502. While when n=1, function_p2 = 7. Hence, as n gets bigger the function will decrease yet still remain positive. In addition, the sum is multiplied by 1/n. Hence, as n gets bigger, the function will get smaller.

_Answer to Problem 3_

```python
def function_p3(n):
    jInterval = np.arange(1,n+1)
    tot = np.sum((jInterval/n)**5. +6.*(jInterval/n)**3.)
    return tot/n
```

5b) (3pts) Fixing $n=10$, $s_{0}=1$, $s_{1}=1$, using your results from Problem 4c, choose three values of $a$, one each from the three different ranges of values of $a$, and generate a plot of $p_{n}$ for each. 5c) (2pts) Fixing $n=10$, $s_{0}=1$, $s_{1}=1$, choose $a=-1/2$ and generate a plot of $p_{n}$.  Briefly explain the results you see and how they differ from those in Problem 5b).  

```python
def fib_general(n, a, s0, s1):
    pn =np.ones(n+1)
    pn[0] =s0
    pn[1] =s1
    if n>=2:
        for nn in range(2, 1+int(n)):
            pn[nn] = pn[nn-1]+ a*pn[nn-2]
    return pn
    
```

```python
plt.plot(fib_general(10,-0.15, 1,1), color = 'b')
plt.plot(fib_general(10,0.15, 1,1), color = 'r')
plt.title("pn")
plt.xlabel("n")
plt.ylabel("$pn$")
plt.legend()
```

```python
plt.plot(fib_general(10,3, 1,1), color = 'g')
plt.title("pn")
plt.xlabel("n")
plt.ylabel("$pn$")
plt.legend()
```
